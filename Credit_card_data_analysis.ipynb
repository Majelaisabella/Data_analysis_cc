```python
n.rename(columns={'default.payment.next.month':'default_pay'},inplace=True)
```

```python
x=n.drop(['default_pay'],axis=1)#feature variable
```

```python
y=n.default_pay#target variable
```

```python
y
```

```python
#split the data in train and test set of data
```

```python
from sklearn.tree import DecisionTreeClassifier#importing decision tree classifier
from sklearn.model_selection import train_test_split#import split function
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
#training the decision tree classification model on the traaining set
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'gini', random_state = 0)
classifier.fit(x_train, y_train)#train tree classification
```

```python
#predicting the new result
```

```python
print(classifier.predict(sc.transform([[8226,20000.0,1,2,33,1,2,2,2,2,2,17971.0,17399.0,19057.0,18453.0,19755.0,19288.0,0.0,2260.0,0.0,1600.0,0.0,644.0]])))
```

```python
classifier= DecisionTreeClassifier()#object for tree classification
classifier.fit(x_train, y_train)#train tree classification
y_pred=classifier.predict(x_test)#predicting the response
```

```python
#evaluvation using accuracy score
```

```python
from sklearn import metrics#for accuracy and confusion matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred)*100)#accuracy 
```

```python
#making the confusion matrix
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
